```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('C:/Users/Aniket/Desktop/new.jpg')
plt.imshow(img)
plt.show()
```

```python
RGB_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
HSV_img=cv2.cvtColor(RGB_img,cv2.COLOR_RGB2HSV)
plt.imshow(HSV_img)
plt.show()
```

```python
lower_yellow=np.array([25,150,50])
upper_yellow=np.array([35,255,255])
mask1=cv2.inRange(HSV_img,lower_yellow,upper_yellow)
real1=cv2.bitwise_and(RGB_img,RGB_img,mask=mask1)
if [mask1]:
    print "YELLOW"
    plt.imshow(real1)
    plt.show()


lower_blue = np.array([110,50,50])
upper_blue = np.array([130,255,255])    
mask2=cv2.inRange(HSV_img,lower_blue,upper_blue)
real2=cv2.bitwise_and(RGB_img,RGB_img,mask=mask2)
if [mask2]:
    print "BLUE"
    plt.imshow(real2)
    plt.show()


lower_red = np.array([0,50,50])
upper_red = np.array([10,255,255])
mask3=cv2.inRange(HSV_img,lower_red,upper_red)
real3=cv2.bitwise_and(RGB_img,RGB_img,mask=mask3)
if [mask3]:
    print "RED"
    plt.imshow(real3)
    plt.show()

    
lower_green= np.array([35,50,50])
upper_green = np.array([75,255,255])    
mask4=cv2.inRange(HSV_img,lower_green,upper_green)
real4=cv2.bitwise_and(RGB_img,RGB_img,mask=mask4)
if [mask4]:
    print "GREEN"
    plt.imshow(real4)
    plt.show()


lower_white= np.array([0,0,0])
upper_white = np.array([0,0,255])    
mask5=cv2.inRange(HSV_img,lower_white,upper_white)
real5=cv2.bitwise_and(RGB_img,RGB_img,mask=mask5)
if [mask5]:
    print "WHITE"
    plt.imshow(real5)
    plt.show()

lower_black= np.array([0,0,0])
upper_black = np.array([180,255,30])    
mask6=cv2.inRange(HSV_img,lower_black,upper_black)
#real6=cv2.bitwise_and(RGB_img,RGB_img,mask=mask6)
if [mask6]:
    print "BLACK"
    plt.imshow(mask6)
    plt.show()
    
```
